```python
import numpy as np 
import pandas as pd

df=pd.read_csv('X_short.csv',index_col=0) # load the dataset
y=pd.read_csv('y_short.csv',index_col=0)
```

```python
#xgboost for feature importance
import xgboost as xgb
from sklearn.metrics import accuracy_score

def get_feature_importance_data(df):
    X = df
    
    train_samples = int(X.shape[0] * 0.7)
 
    X_train = X.iloc[:train_samples]
    X_test = X.iloc[train_samples:]

    y_train = y.iloc[:train_samples]
    y_test = y.iloc[train_samples:]
    
    return (X_train, y_train), (X_test, y_test)
```

```python
(X_train_FI, y_train_FI), (X_test_FI, y_test_FI) = get_feature_importance_data(df)
```

```python
regressor = xgb.XGBRegressor(gamma=0.0,n_estimators=150,base_score=0.7,colsample_bytree=1,learning_rate=0.05)
```

```python
xgbModel = regressor.fit(X_train_FI,y_train_FI, \
                         eval_set = [(X_train_FI, y_train_FI), (X_test_FI, y_test_FI)], \
                         verbose=True)
```

```python
eval_result = regressor.evals_result()
```

```python
training_rounds = range(len(eval_result['validation_0']['rmse']))
```

```python
#Recursive Feature Elimination with Logistic Regression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


# create a base classifier used to evaluate a subset of attributes
model = LogisticRegression()
# create the RFE model and select 3 attributes
rfe = RFE(model, 3)
rfe = rfe.fit(X_train, y_train)

y_pred = rfe.predict(X_test)

print(confusion_matrix(y_test,y_pred)) 
print(classification_report(y_test, y_pred))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# summarize the selection of the attributes
print(rfe.support_)
print(rfe.ranking_)
#   . 1 -> , True   features  
```

```python
#feature importance

import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(8,8))
plt.xticks(rotation='vertical')
plt.bar([i for i in range(len(rfe.ranking_))], rfe.ranking_.tolist(), tick_label=X_test.columns)
plt.title('Figure 6: Feature importance of the technical indicators.')
plt.show()

print(dict(zip(X_train.columns, rfe.ranking_)))
```

```python
#Extra Trees ensemble

from sklearn.ensemble import ExtraTreesClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


# fit an Extra Trees model to the data
model = ExtraTreesClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(confusion_matrix(y_test,y_pred)) 
print(classification_report(y_test, y_pred))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# display the relative importance of each attribute
print(model.feature_importances_)
```

```python
#random forest

from pandas import read_csv
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

np.random.seed(1)

trees=250
max_feat=7
max_depth=30
min_sample=2

clf = RandomForestClassifier(n_estimators=trees,max_features=max_feat,max_depth=max_depth,min_samples_split= min_sample, random_state=0,n_jobs=-1)

#Train the classifier and calculate the training time

clf.fit(X_train, y_train) 

y_pred = clf.predict(X_test)

print(confusion_matrix(y_test,y_pred)) 
print(classification_report(y_test, y_pred))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#LASSO

from sklearn.linear_model import lasso_path, enet_path

_, coef_path, _ = lasso_path(X_train, y_train, alphas=[1.])
print(coef_path)
```

```python
weight = [None]*26

for i in range(26):
    variable = np.mean(coef_path[:,i,:])
    weight[i] = variable
    
final_weight = np.asarray(weight)
final_weight = final_weight.reshape(26,1)
```
